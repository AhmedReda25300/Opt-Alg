```python
def single_var_GD(X,y,lr,epochs,batch_size):
    np.random.seed(101)
    data = np.concatenate((X.reshape(-1,1),y.reshape(-1,1)),axis=1)
    np.random.shuffle(data)
    X = data[:,0]
    y = data[:,1]
    w=0
    b=0
    losses=[]
    W=[]
    B=[]
    for epoch in range(epochs):
        for i in range(0,len(X),batch_size):
            # X=X[i:i+batch_size]
            # y=y[i:i+batch_size]
            hx=X[i:i+batch_size]*w+b
            error=hx-y[i:i+batch_size]
            loss=1/2 * np.sum(error**2) /len(X[i:i+batch_size])
            dw=np.mean(error*X[i:i+batch_size])
            db=np.mean(error)
            GD_V=np.array([dw,db])
            GD_Norm=np.linalg.norm(GD_V)
            #update
            w=w-lr*GD_V[0]
            b=b-lr*GD_V[1]
        #store
        losses.append(loss)
        W.append(w)
        B.append(b)
        #print
        print('_____________________________________________________________')
        print(f"epoch:{epoch} loss:{loss} w:{w} b:{b} GD_V:{GD_V} GD_Norm:{GD_Norm}")
        if GD_Norm<0.001 or abs(losses[epoch]-losses[epoch-1])<0.001 and epoch != 0:
            print("converged")
            print(f"epoch:{epoch} loss:{loss} w:{w} b:{b} GD_V:{GD_V}")
            break
    return losses,W,B
        
            
```

```python
def single_var_NAG_GD(X, y, lr, epochs, batch_size, momentum):
    np.random.seed(101)
    data = np.concatenate((X.reshape(-1,1), y.reshape(-1,1)), axis=1)
    np.random.shuffle(data)
    X = data[:,0]
    y = data[:,1]
    w = 0
    b = 0
    losses = []
    W = []
    B = []
    Vw = 0  # Initialize velocity for weights
    Vb = 0  # Initialize velocity for bias
    for epoch in range(epochs):
        for i in range(0, len(X), batch_size):
            # Compute gradients using current momentum-based update
            w_temp = w - momentum * Vw
            b_temp = b - momentum * Vb
            hx_temp = X[i:i+batch_size] * w_temp + b_temp
            error = hx_temp - y[i:i+batch_size]
            dw_temp = np.mean(error * X[i:i+batch_size])
            db_temp = np.mean(error)

            # Update velocities with the computed gradients
            Vw = momentum * Vw + lr * dw_temp
            Vb = momentum * Vb + lr * db_temp

            # Update weights and bias
            w = w - Vw
            b = b - Vb

            # Compute loss
            loss = 0.5 * np.mean(error**2)

            # Compute gradient norm
            GD_Norm = np.linalg.norm(np.array([dw_temp, db_temp]))

        # Store losses and parameters
        losses.append(loss)
        W.append(w)
        B.append(b)

        # Print epoch details
        print('_____________________________________________________________')
        print(f"Epoch:{epoch} Loss:{loss} w:{w} b:{b} dw:{dw_temp} db:{db_temp} GD_Norm:{GD_Norm}")

        # Convergence check
        if GD_Norm < 0.001 or (epoch != 0 and abs(losses[epoch] - losses[epoch-1]) < 0.001):
            print("Converged")
            print(f"Epoch:{epoch} Loss:{loss} w:{w} b:{b} dw:{dw_temp} db:{db_temp} GD_Norm:{GD_Norm}")
            break
 
```

```python
def multi_var_NAG_GD(X, y, lr, epochs, batch_size, momentum):
    np.random.seed(101)
    data = np.concatenate((X, y.reshape(-1, 1)), axis=1)
    np.random.shuffle(data)
    # m, n = data.shape  # Get number of samples (m) and features (n)

    # Add a column of ones for the bias term and separate features and target
    X_shuffled = np.concatenate((np.ones((m, 1)), data[:, :n-1]), axis=1)
    y_shuffled = data[:, n-1].reshape(-1, 1)

    w = np.zeros((n, 1))  # Initialize weights
    Vw = np.zeros((n, 1))  # Initialize velocities

    W = []  # List to store weights at each epoch
    losses = []  # List to store losses at each epoch

    for epoch in range(epochs):
        for i in range(0, m, batch_size):
            hx = X_shuffled[i:i+batch_size] @ w
            error = hx - y_shuffled[i:i+batch_size]
            loss = 1/2 * np.sum(error**2) /len(X_shuffled[i:i+batch_size])

            dw = (X_shuffled[i:i+batch_size].T @ error) / len(X_shuffled[i:i+batch_size])

            # Update velocities
            w_temp = w - momentum * Vw
            hx_temp = X_shuffled[i:i+batch_size] @ w_temp
            error_temp = hx_temp - y_shuffled[i:i+batch_size]
            
            dw_temp = (X_shuffled[i:i+batch_size].T @ error_temp) / len(X_shuffled[i:i+batch_size])
            Vw = momentum * Vw + lr * dw_temp

            # Update weights using the velocity
            w = w_temp - lr * dw_temp

            # Compute gradient norm
            GD_Norm = np.linalg.norm(dw)

        # Store losses and parameters
        losses.append(loss)
        W.append(w)

        # Print epoch details
        print('_____________________________________________________________')
        print(f"epoch:{epoch} loss:{loss} w:{w} GD_V:{dw_temp} GD_Norm:{GD_Norm}")
        if GD_Norm < 0.001 or (epoch != 0 and abs(losses[epoch] - losses[epoch-1]) < 0.001):
            print("converged")
            print(f"epoch:{epoch} loss:{loss} w:{w} GD_V:{dw_temp} GD_Norm:{GD_Norm}")
            break

    return losses, W
```

```python
def single_variable_Adagrad(X,y,lr,epochs,batch_size,epsilon):
    np.random.seed(101)
    data = np.concatenate((X.reshape(-1,1),y.reshape(-1,1)),axis=1)
    np.random.shuffle(data)
    x_shuffled =data[:,0]
    y_shuffled = data[:,1]
    w=0
    b=0
    Vw=0
    Vb=0
    W= []
    losses= []
    for epoch in range(epochs):
        for i in range(0,len(x_shuffled),batch_size):
            hx=x_shuffled[i:i+batch_size]*w+b
            error=hx-y_shuffled[i:i+batch_size]
            loss=1/2 * np.sum(error**2) /len(x_shuffled[i:i+batch_size])
            dw=np.mean(error*x_shuffled[i:i+batch_size])
            db=np.mean(error)
            GD_V=np.array([dw,db])
            GD_Norm=np.linalg.norm(GD_V)
            #update
            Vw=Vw+GD_V[0]**2
            Vb=Vb+GD_V[1]**2
            w=w-(lr/np.sqrt(Vw+epsilon))*GD_V[0]
            b=b-(lr/np.sqrt(Vb+epsilon))*GD_V[1]
        losses.append(loss)
        W.append(w)
        B.append(b)
        print('_____________________________________________________________')
        print(f"epoch:{epoch} loss:{loss} w:{w} GD_V:{GD_V} GD_Norm:{GD_Norm}")
        if GD_Norm < 0.001 or (epoch != 0 and abs(losses[epoch] - losses[epoch-1]) < 0.001):
            print("converged")
            print(f"epoch:{epoch} loss:{loss} w:{w} GD_V:{GD_V} GD_Norm:{GD_Norm}")
            break


    return losses, W
```

multi variable Adagrad

```python
def multi_var_Adagrad(X, y, lr, epochs, batch_size, epsilon):
    np.random.seed(101)
    data = np.concatenate((X, y.reshape(-1, 1)), axis=1)
    np.random.shuffle(data)
    m, n = data.shape  
    X_shuffled = np.concatenate((np.ones((m, 1)), data[:, :n-1]), axis=1)
    y_shuffled = data[:, n-1].reshape(-1, 1)

    w = np.zeros((n, 1))  
    Vw = np.zeros((n, 1)) 
    W = []
    losses = []

    for epoch in range(epochs):
        for i in range(0, len(X_shuffled), batch_size):
            hx = X_shuffled[i:i+batch_size] @ w
            error = hx - y_shuffled[i:i+batch_size]
            loss = 0.5 * np.sum(error**2) / len(X_shuffled[i:i+batch_size])
            dw = (X_shuffled[i:i+batch_size].T @ error) / len(X_shuffled[i:i+batch_size])
            GD_Norm = np.linalg.norm(dw)

            # Update
            Vw = Vw + dw**2
            w = w - lr * (1 / (np.sqrt(Vw) + epsilon)) * dw

        losses.append(loss)
        W.append(w)
        
        # Print epoch details
        print('_____________________________________________________________')
        print(f"epoch:{epoch} loss:{loss} w:{w} GD_V:{dw} GD_Norm:{GD_Norm}")
        
        # Convergence criterion
        if GD_Norm < 0.001 or (epoch != 0 and abs(losses[epoch] - losses[epoch-1]) < 0.001):
            print("converged")
            print(f"epoch:{epoch} loss:{loss} w:{w} GD_V:{dw} GD_Norm:{GD_Norm}")
            break

    return losses, W
```

```python
def single_var_rms_prop(X,y,lr,epochs,batch_size,epsilon,beta):
    np.random.seed(101)
    data = np.concatenate((X.reshape(-1,1),y.reshape(-1,1)),axis=1)
    np.random.shuffle(data)
    x_shuffled =data[:,0]
    y_shuffled = data[:,1]
    w=0
    b=0
    Vw=0
    Vb=0
    W= []
    B= []
    losses= []
    for epoch in range(epochs):
        for i in range(0,len(x_shuffled),batch_size):
            hx=x_shuffled[i:i+batch_size]*w+b
            error=hx-y_shuffled[i:i+batch_size]
            loss=1/2 * np.sum(error**2) /len(x_shuffled[i:i+batch_size])
            db=np.sum(error) / len(x_shuffled[i:i+batch_size])
            dw=np.sum(error * x_shuffled[i:i+batch_size]) / len(x_shuffled[i:i+batch_size])
            GD_V=np.array([dw,db])
            GD_Norm=np.linalg.norm(GD_V)
            #update
            Vw=beta * Vw + (1-beta) * (GD_V[0]**2)
            Vb=beta * Vb + (1-beta) * (GD_V[1]**2)
            w = w - lr * (1/(np.sqrt(Vw) + epsilon)) * GD_V[0]
            b= b - lr * (1/(np.sqrt(Vb) + epsilon)) * GD_V[1]
        losses.append(loss)
        W.append(w)
        B.append(b)
        print('_____________________________________________________________')
        print(f"epoch:{epoch} loss:{loss} w:{w} ,b:{b} GD_V:{GD_V} GD_Norm:{GD_Norm}")
        if GD_Norm < 0.001 or (epoch != 0 and abs(losses[epoch] - losses[epoch-1]) < 0.001):
            print("converged")
            print(f"epoch:{epoch} loss:{loss} w:{w} ,b:{b} GD_V:{GD_V} GD_Norm:{GD_Norm}")
            break

    return losses, W
```
