```python
import numpy as np
```

```python
X=np.linspace(0,20)
a=-1
b=2
y=a*X+b 
```

```python
def single_variable_GD(X,y,learning_rate,epochs,bathch_size):
    data=np.concatenate((X.reshape(-1,1),y.reshape(-1,1)),axis=1)
    np.random.shuffle(data)
    X_shuffled=data[:,0]
    y_shuffled=data[:,1]

    thetas_0=[]
    thetas_1=[]
    theta_0=0
    theta_1=0

    epoch_loss=[]
    loss=[]
    m=len(X_shuffled)

    for epoch in range(epochs):
        for i in range(0,m,bathch_size):
            h_x=theta_0+theta_1*X_shuffled[i:i+bathch_size]
            error=h_x-y_shuffled[i:i+bathch_size]
            j=1/2 * np.sum(error**2) /len(X_shuffled[i:i+bathch_size])

            thetas_0.append(theta_0)
            thetas_1.append(theta_1)
            loss.append(j)

            d_theta_0=np.sum(error)/len(y_shuffled[i:i+bathch_size])
            d_theta_1=np.sum(error*X_shuffled[i:i+bathch_size])/len(y_shuffled[i:i+bathch_size])

            gradient_vector=np.array([[d_theta_0],[d_theta_1]])
            Norm=np.linalg.norm(gradient_vector)
            theta_0=theta_0-learning_rate*d_theta_0
            theta_1=theta_1-learning_rate*d_theta_1
        epoch_loss.append(j)

        print("****************** Epoch ",epoch,"******************")
        print("j =",j,"\n")
        print("Gradient Vector:\n",gradient_vector,"\n")
        print("Gradiant Vector Norm:\n",Norm,"\n")
        print("theta_0_new:",theta_0)
        print("theta_1_new:",theta_1)

        if Norm <0.001 or abs(epoch_loss[epoch]-epoch_loss[epoch-1])<0.001 and epoch!=0:
            print("****************** Training Report ******************\n")
            print("Gradient Descent converged after ",epoch," epochs\n")
            print("theta_0_Opt:",theta_0)
            print("theta_1_Opt:",theta_1,"\n")
            print("Cost = ",j,"\n")
            break

    return thetas_0,thetas_1,loss,theta_0,theta_1,X_shuffled,y_shuffled
```

```python
single_variable_GD(X,y,0.0005,300,len(X))
```
